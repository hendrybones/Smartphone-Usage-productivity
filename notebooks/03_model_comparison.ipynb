#### Importing Functions

```python
# Core imports
import sys
from pathlib import Path
import pandas as pd

# Add your script folder so we can import functions
sys.path.append("../src/models")

# Project root
project_root = Path(r"C:\Users\Administrator\OneDrive\Documents\Data Science\Smartphone-Usage-productivity")

# Import your pipeline functions
from model_comparison import load_config, load_data, prepare_data, compare_models, get_models, evaluate_model
```

### Load config and data

```python

config_path = project_root / "config" / "config.yaml"
config = load_config(config_path)

# Ensure data path is absolute
config['data']['processed_path'] = str(project_root / "data" / "processed")

# Load dataset
df = load_data(config)
```

### Prepare the data (train/test split + preprocessing)

```python
X_train, X_test, y_train, y_test = prepare_data(df, config)
```

### Train your models

```python
# Get all models
models = get_models()

# Train each model and store results
trained_models = {}
for name, model in models.items():
    model.fit(X_train, y_train)
    trained_models[name] = model
    print(f"{name} trained.")
```

### Evaluate models on test set

```python
results = []
for name, model in trained_models.items():
    metrics = evaluate_model(model, X_train, y_train, X_test, y_test)
    results.append({"Model": name, **metrics})

results_df = pd.DataFrame(results).sort_values(by="R2 Score", ascending=False)
results_df
```
